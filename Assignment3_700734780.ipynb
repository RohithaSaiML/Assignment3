```python
# split into X and Y
X = train.drop(columns = ['Survived','Name']).values
y = train.Survived.values

print(X.shape)
print(y.shape)
```

```python
#to split data into train and test dataset(75-25)
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(X, y, test_size = .25, random_state = 0, stratify = y)
```

```python
from collections import Counter
def get_ouliers(data):
    outliers = []
    
    # to iterate through the columns
    for col in data.columns.tolist():
        #first 25% of data
        Q1 = np.percentile(data[col], 25)
        
        #next 75% of data
        Q3 = np.percentile(data[col],75)

        # Interquartile range
        IQR = Q3 - Q1
        
        # to find the outlier range
        outlier_range = 1.5 * IQR
        
        # to determine the outliers for column
        outliers_values = data[(data[col] < Q1 - outlier_range) | (data[col] > Q3 + outlier_range)].index
        
        # appending the outlier indices for col to the list of outlier indices 
        outliers.extend(outliers_values)
        
    #to select columns with more than two outliers
    outliers = Counter(outliers)
    multiple_outliers = list( k for k, v in outliers.items() if v > 3 )
    
    return multiple_outliers   

print('There are %d observations having more than 2 outliers' %(len(get_ouliers(df[features]))))   
```

```python
#to remove the outliers
outliers_index = get_ouliers(df[features])
df = df.drop(outliers_index)
```

```python
# to split data to features and target
X = df.drop(columns = ['Type']).values
y = df.Type.values

print(X.shape)
print(y.shape)
```

```python
# to split data into 75 train and 25 test
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(X, y, test_size = .25, random_state = 0, stratify = y)
```

```python
# to scale our features
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
#2.1 Naive Bayes
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)

#2.1 Support Vector Machine
from sklearn.svm import SVC
svm = SVC(kernel = 'rbf')
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Naive Bayes
print("Naive Bayes Confusion Matrix:\n", confusion_matrix(y_test, nb_pred))

#2.2 Classification report on testing part using score on Naive bayes
print("Classification Report for Naive Bayes:\n", classification_report(y_test, nb_pred))
print()

# Support Vector Machine
print("Support Vector Machine Confusion Matrix:\n", confusion_matrix(y_test, svm_pred))

#2.2 Classification report on testing part using score on SVM
print("Classification Report for SVM:\n", classification_report(y_test, svm_pred))
print()
```

```python
#Accuracy is better in SVM than Naive bayes
svm = accuracy_score(y_test,svm_pred)
print('Accuracy for svm: ',svm)
```

```python
nb = accuracy_score(y_test,nb_pred)
print('Accuracy in Naive Bayes: ',nb)
```
